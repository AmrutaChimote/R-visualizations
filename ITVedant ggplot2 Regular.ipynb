```r
##ggplot
#grammer of graphics

#Data - dataset
#Aesthetics - x,y,color, fill,size, labels,alpha,shape-aes()
#Geometrics: point,line,histogram,bar,boxplot-geom()
#facets: subset of data using variuos rows and columns
#statistics: binning, smoothing,descriptive
#coordinates : limits, fixed
#themes 
```

```r
install.packages("ggplot2")
```

```r
library("ggplot2")
```

```r
data(mtcars)
head(mtcars)
```

```r
#data layer
ggplot(data=mtcars)
```

```r
#Aesthetic layer
ggplot(data=mtcars,aes(x=hp,y=mpg))
```

```r
#geometric layer
ggplot(data=mtcars,aes(x=hp,y=mpg))+geom_point()
```

```r
#geometric layer
ggplot(data=mtcars,aes(x=hp,y=mpg,col=am))+geom_point()
```

```r
ggplot(data=mtcars,aes(x=hp,y=mpg,col=cyl))+geom_point()
```

```r
ggplot(data=mtcars,aes(x=hp,y=mpg,size=disp,col=cyl))+geom_point()
```

```r
ggplot(data=mtcars,aes(x=hp,y=mpg,col=factor(cyl)))+geom_point()
```

```r
ggplot(data=mtcars,aes(x=hp,y=mpg,col=factor(cyl),shape=factor(am)))+geom_point()
```

```r
a<- ggplot(data=mtcars,aes(x=hp,y=mpg,shape=factor(cyl),col=factor(am)))+geom_point()
a
```

```r
a+facet_grid(.~cyl)#.after dot column wise (before column name ~)
#before dot . row wise (atfer row name ~)
```

```r
a+facet_grid(am~.)#separate rows based on transmission am
```

```r
a+facet_grid(.~cyl)#sep col based on cylinders
```
